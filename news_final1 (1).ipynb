```python
import requests
import pandas as pd
import json
import re 
```

```python
total=[]
for k in range(14,20):
    for i in range(1, 13):
        link='https://www.rbc.ru/v10/search/ajax/?project=rbcnews&dateFrom=01.' + str(i) + '.20' + str(k) + '&dateTo=31.' + str(i) + '.20' + str(k) + '&offset=0&limit=1000&query=%D1%81%D0%B0%D0%BD%D0%BA%D1%86%D0%B8%D0%B8'
        table=requests.get(link).content.decode('utf-8')
        page4 = re.sub('..counts.*', '', table)+'}'
        total.append(pd.DataFrame(json.loads(page4)['items']))
```

```python
news_total = pd.concat(total)
```
