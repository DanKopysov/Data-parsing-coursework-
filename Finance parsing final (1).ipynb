```python
import json
import pandas as pd
import requests
import re
import itertools
```

```python
#Табличка со всеми компаниями которые есть
```

```python
i=0
total = []
for i in range(0, 301, 100):
    if i<=300:
        t = 'https://iss.moex.com/iss/history/engines/stock/markets/shares/boardgroups/57/securities.jsonp?lang=ru&security_collection=3&date=2016-06-07&start=' + str(i) + '&iss.meta=off&iss.json=extended&callback=angular.callbacks._g&sort_column=VALUE&sort_order=desc&limit=100&_=1557332537830'
        page = requests.get(t).content.decode('utf-8')
        page = re.sub('\n', '', page)
        page2 = page[57:]
        page4 = re.sub('..history.cursor.*', '', page2)+'}'
        total.append(pd.DataFrame(json.loads(page4)['history']))
    else:
        break
```

```python
result = pd.concat(total)
```

```python
result
```

```python
len(result.SHORTNAME.unique())
```

```python
result.to_excel("Stock_companies.xlsx", sheet_name='Sheet_name_1')
```

```python
#Табличка с выбранными компаниями
```

```python
tickers = ['UNAC', 'ROSN', 'VTBR', 'TRNFP', 'SIBN', 'SBER', 'NVTK', 'GAZP', 'LKOH', 'MSTT', 'SNGS', 'CBOM', 'MMBM', 'ROSB', 'BSPB', 
'LSRG', 'RKKE', 'TATN', 'BANE', 'RNFT', 'MFGS', 'VJGZ', 'YAKG', 'KRKN']
```

```python
companies = []
for k in tickers:
    for l in range(0, 1301, 100):
        link = 'https://iss.moex.com/iss/history/engines/stock/markets/shares/boardgroups/57/securities/' + str(k) + '.jsonp?lang=ru&from=2014-01-01&till=2019-05-10&start=' + str(l) + '&iss.meta=off&iss.json=extended&callback=angular.callbacks._9&sort_order=TRADEDATE&sort_order_desc=desc&limit=100&_=1557839939152'
        table = requests.get(link).content.decode('utf-8')
        table = re.sub('\n', '', table)
        table2 = table[57:]
        table4 = re.sub('..history.cursor.*', '', table2)+'}'
        companies.append(pd.DataFrame(json.loads(table4)['history']))
```

```python
len(companies)
```

```python
result_total_companies = pd.concat(companies)
```
